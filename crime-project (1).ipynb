```python
for col in d4.columns:
    if d4.skew().loc[col]>0.55:
        d4[col]=np.log1p(d4[col])
d4.skew()
```

```python
from scipy.stats import zscore
z=np.abs(zscore(d4))
threshold=3
print(np.where(z>3))
```

```python
d4_new=d4[(z<3).all(axis=1)]
print(d4.shape)
print(d4_new.shape)
```
